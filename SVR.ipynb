```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd
import numpy as np
```

```python
data_from_cst=pd.read_csv(r"C:\Jupyter Demo\optimized_antenna_dataset_v2.csv")
data_from_cst
```

```python
x=data_from_cst.iloc[:,:-1]
y=data_from_cst.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
regress=Pipeline([('scaler', StandardScaler()),
                 ('svr', SVR(kernel='rbf',C=500,epsilon=0.001,gamma='auto'))])
regress.fit(x_train,y_train)
```

```python
y_pred=regress.predict(x_test)
```

```python
#RMSE
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
#MSE
print(metrics.mean_squared_error(y_test,y_pred))
```

```python
#MAE
from sklearn import metrics
print(metrics.mean_absolute_error(y_pred,y_test))
```

```python
#R2 score
print(metrics.r2_score(y_pred,y_test))
```

```python
#variance
print(metrics.explained_variance_score(y_pred,y_test))
```
